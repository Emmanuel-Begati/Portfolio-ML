### 3. **Data Exploration**
- **Features Included:**
  - `FEN`: The board state in Forsyth-Edwards Notation.
  - `Moves`: Solution moves for the puzzle.
  - `Rating`: Difficulty rating of the puzzle.
  - `RatingDeviation`: Variability in the puzzle rating.

  - `Evaluation`: Engine evaluation score (in centipawns).

  - `black_material`: Total material value for Black.
  - `white_pieces`: List of pieces White has on the board.
  - `black_pieces`: List of pieces Black has on the board.



### Summary of Findings:
- **Distributions:** Puzzle ratings show a normal distribution with a peak around the average rating value which is between 1000 and 1700.
- **Correlations:** High correlation between `white_material` and `black_material`, suggesting material balance is often maintained in puzzles. `Evaluation` has a moderate negative correlation with `black_material`, indicating higher evaluations favor White.
- **Outliers:** Outliers were identified in features like `NbPlays` and `Evaluation`, which may need handling before model training.
- **Themes Analysis:** The most common puzzle themes include basic tactical motifs like forks, pins, and discovered attacks.**


```python
# Display basic information
print(df.info())
print(df.describe())

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Convert numeric columns if needed
numeric_columns = ['Rating', 'RatingDeviation', 'Popularity', 'NbPlays', 'Evaluation', 'white_material', 'black_material']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Distribution of puzzle ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=50, kde=True)
plt.title('Distribution of Puzzle Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Correlation heatmap (numeric features only)
numeric_data = df[numeric_columns]
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

# Scatter plot of NbPlays vs Popularity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NbPlays', y='Popularity', data=df)
plt.title('NbPlays vs Popularity')
plt.xlabel('Number of Plays')
plt.ylabel('Popularity')
plt.show()

# Analysis of puzzle themes
theme_counts = df['Themes'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=theme_counts.index, y=theme_counts.values)
plt.title('Top 10 Most Common Puzzle Themes')
plt.xticks(rotation=45)
plt.xlabel('Theme')
plt.ylabel('Count')
plt.show()

# Outliers detection using boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_data)
plt.title('Boxplot of Numeric Features to Identify Outliers')
plt.xticks(rotation=45)
plt.show()
```

### 5. **Handling Sparse/Dense Data and Outliers:**
   - Assess the density of the data.
   - Propose strategies to handle missing data and outliers while maintaining dataset integrity.

   **Insert code for Handling Sparse/Dense Data and Outliers below**

```python
# Assess data density (missing values)
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("Missing Values (%):\n", missing_percentage)

# Drop columns with more than 30% missing values (if any)
sparse_columns = missing_percentage[missing_percentage > 30].index
df.drop(columns=sparse_columns, inplace=True)
print(f"Dropped columns due to high sparsity: {sparse_columns}")

# Handling missing data for numeric columns (fill with median)
numeric_columns = ['Rating', 'RatingDeviation', 'Popularity', 'NbPlays', 'Evaluation', 'white_material', 'black_material']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

# Handling missing data for categorical columns (fill with mode)
df['Themes'].fillna(df['Themes'].mode()[0], inplace=True)

# Detect and handle outliers using the IQR method
def handle_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Apply outlier handling to numeric columns
for col in numeric_columns:
    handle_outliers(col)

# Visualize outliers after handling
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_columns])
plt.title('Boxplot of Numeric Features After Outlier Handling')
plt.xticks(rotation=45)
plt.show()

# Recheck missing values
print("Missing Values After Handling (%):\n", df.isnull().sum() / len(df) * 100)
```
